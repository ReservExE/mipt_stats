```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')
df.info()
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x=df['Date'], y=df['Close'])

plt.title('Stock price dynamics')
plt.xlabel('Date')
plt.ylabel('Close price')
plt.show()
```

Проверим является ряд выше стационарным
Проведем тест дики-фуллера

```python
from statsmodels.tsa.stattools import adfuller
adfuller(df['Close'])
```

p-value > 0.05 => не можем отвергать H0 о нестационарности ряда
